```python
d_vocab = {}
d = {}
```

```python
path = "./question/total_question.txt"
```

```python
count = 0
with open(path,"r") as fr:
    lines = fr.readlines()
    for line in lines:
        #print(line.strip())
        line = line.strip()
        terms = HanLP.segment(line)
        for term in terms:
            print(term.word)
            if(str(term).split('/')[-1]!='w'):
                if(term.word not in d_vocab):
                    d_vocab[term.word]=count
                    d[count]=term.word
                    count=count+1;
```
